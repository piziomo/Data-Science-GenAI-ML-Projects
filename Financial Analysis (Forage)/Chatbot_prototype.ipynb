```python
#Load necessary libraries and the financial data
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('/content/sample_data/Data for BCG/top_companies_earnings.csv')

# Quick look at the data
df.head()
```

```python
# define queries

def simple_chatbot(user_query):
    """Responds to 5 predefined financial queries."""
    q = user_query.lower().strip()

    if q == "what is the total revenue?":
        return f"The total revenue across all companies in {chatbot_data['latest_year']} is {chatbot_data['total_revenue']:,} million USD."

    elif q == "how has apple's net income changed over the last year?":
        direction = "increased" if chatbot_data['apple_income_change'] > 0 else "decreased"
        return f"Apple's net income has {direction} by {abs(chatbot_data['apple_income_change']):,.0f} million USD from 2023 to 2024."

    elif q == "who had the highest cash flow in 2024?":
        return f"{chatbot_data['top_cash_company']} had the highest cash flow in 2024: {chatbot_data['top_cash_value']:,} million USD."

    elif q == "show tesla's revenue trend":
        return chatbot_data['tesla_trend'].to_string(index=False)

    elif q == "compare microsoft and apple's revenue in 2024":
        return f"In 2024, Microsoft had a revenue of {chatbot_data['ms_revenue']:,} million USD and Apple had {chatbot_data['ap_revenue']:,} million USD."

    else:
        return "Sorry, I can only provide information on predefined queries."
```

```python
#chatbot response from data

def prepare_chatbot_data(df):
    """Prepares necessary values for the chatbot."""
    latest_year = df['Year'].max()
    latest_df = df[df['Year'] == latest_year]

    # Apple's net income change
    apple_income_2024 = df[(df['Company_name'] == 'Apple') & (df['Year'] == 2024)]['net_income'].values[0]
    apple_income_2023 = df[(df['Company_name'] == 'Apple') & (df['Year'] == 2023)]['net_income'].values[0]
    net_income_change = apple_income_2024 - apple_income_2023

    # Tesla revenue trend
    tesla_trend = df[df['Company_name'] == 'Tesla'][['Year', 'total_revenue']]

    # Highest cash flow in 2024
    top_cash_2024 = df[df['Year'] == 2024].sort_values(by='cash_flow', ascending=False).iloc[0]

    # Microsoft and Apple revenue comparison
    ms_revenue = df[(df['Company_name'] == 'Microsoft') & (df['Year'] == 2024)]['total_revenue'].values[0]
    ap_revenue = df[(df['Company_name'] == 'Apple') & (df['Year'] == 2024)]['total_revenue'].values[0]

    return {
        "latest_year": latest_year,
        "total_revenue": latest_df['total_revenue'].sum(),
        "apple_income_change": net_income_change,
        "top_cash_company": top_cash_2024['Company_name'],
        "top_cash_value": top_cash_2024['cash_flow'],
        "tesla_trend": tesla_trend,
        "ms_revenue": ms_revenue,
        "ap_revenue": ap_revenue
    }

# Store extracted values
chatbot_data = prepare_chatbot_data(df)

```

```python
# Test the chatbot
if __name__ == "__main__":
    test_queries = [
        "What is the total revenue?",
        "How has Apple's net income changed over the last year?",
        "Who had the highest cash flow in 2024?",
        "Show Tesla's revenue trend",
        "Compare Microsoft and Apple's revenue in 2024",
        "What is Tesla's profit margin?"  # Not predefined
    ]

    for query in test_queries:
        print(f"\nUser: {query}")
        print(f"Bot: {simple_chatbot(query)}")
```
